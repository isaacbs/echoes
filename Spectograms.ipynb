# Spectograms of .Wav files


```python
# Import necessary libraries
import librosa
import librosa.display as ld
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load in sample
cantina, rate = librosa.load('./slices/cantina.wav')
```

```python
fourier = np.abs(librosa.stft(cantina))
decibles = librosa.amplitude_to_db(fourier, ref = np.max)
```

```python
fig, ax = plt.subplots(figsize=(25, 10))

specto = ld.specshow(decibles, sr=rate, x_axis="time", y_axis="linear", cmap='plasma', ax=ax)
ax.set(title="Cantina Spectogram")
fig.colorbar(specto)
```

## Mel Scale and Mel Spectograms

A Mel Spectogram is different from a normal spectogram because of the scale that it is placed on. The Mel scale is related to the Hertz scale with the equation: m = 1127 * ln(1+ frequency/700)

```python
def mel_specto(sound):
    spec = librosa.feature.melspectrogram(sound)
    dec_spec = librosa.amplitude_to_db(spec)
    return dec_spec
```

```python
fig, ax = plt.subplots(figsize=(25, 10), dpi=200)
a = ld.specshow(mel_specto(cantina), x_axis='time', ax=ax, cmap='plasma')
plt.colorbar(a)
ax.set_title("Mel Spectogram of Cantina")
```

### MFCC (Mel Frequency Cepstral Coefficients)

MFCCs are Mel Frequency Cepstral Coefficients which are able to represent the timber of different instruments well.

```python
cantinaMFCC = librosa.feature.mfcc(cantina)
```

```python
fig, ax = plt.subplots(figsize=(25,10), dpi=200)
a = ld.specshow(cantinaMFCC, ax=ax, x_axis='time', cmap='plasma')
ax.set_title('Cantina MFCC')
plt.colorbar(a)
```
