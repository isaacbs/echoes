# Fourier Transformation for wav data

## We will be using the Fast Fourier Transformation function from numpy in order to transform our data into its individual frequencies.

This will allow for us to understand which frequencies are more prevalent within a certain sound and will lead us towards being able to create spectograms.

```python
# Imports for necessary libraries
import librosa
import librosa.display as ld
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Load .wav file
cantina, rate = librosa.load('./slices/cantina.wav')
```

Define a fast fourier tranformation function which will return the magnitude of the waves and a linearly spaced out scale of the frequency from highest to lowest.

```python
def fast_fourier(sig):
    fourier = np.fft.fft(sig)
    magnitude = np.abs(np.real(fourier))
    frequency = np.linspace(0, rate, len(magnitude))
    return magnitude, frequency
```

```python
canMAG, canFREQ = fast_fourier(cantina)
```

```python
fig, ax = plt.subplots(1, figsize=(12,8), dpi = 200)
ax.plot(canFREQ, canMAG)
ax.set(title="Fast Fourier Transform")
```

We can see that the graph is symmetrical, so we need only to look at one half of it.

```python
canHalf = int(len(canMAG)/2)
fig, ax = plt.subplots(1, figsize=(12,8), dpi=200)
ax.plot(canFREQ[:canHalf], canMAG[:canHalf])
ax.set(title="FFT on Cantina")
```

This gives us a better way of understanding which frequencies are most common in our slice of sound.
