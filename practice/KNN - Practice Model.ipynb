```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Data/sonar.all-data.csv')
```

```python
df
```

Heat Map showing correlation between frequencies

```python
fig, ax = plt.subplots(figsize=(20,12), dpi =200)
sns.heatmap(df.corr(), ax=ax, cmap='coolwarm')
```

```python
df['Target'] = df['Label'].map({'R': 0, 'M': 1})
```

```python
df
```

```python
np.abs(df.corr()['Target']).sort_values().tail()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['Target','Label'], axis=1)
```

```python
y = df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
scaler = StandardScaler()
knn = KNeighborsClassifier()
```

```python
ops = [('scaler', scaler),('knn',knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(ops)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values = list(range(1,20))
```

```python
param_grid = {'knn__n_neighbors':k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train, y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
dd = pd.DataFrame(full_cv_classifier.cv_results_)
```

```python
dd
```

```python
fig, ax = plt.subplots(figsize = (20,12), dpi=200)
plt.plot(dd['mean_test_score'])
```

```python
y_pred = full_cv_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
