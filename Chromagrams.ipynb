# Chromagrams
## Separating our data out into distinctive notes
Chromagrams allow for us to visualize sounds as specific pitches along the musical scale. Which consists of 12 notes, C-C#-D-D#-E-F-F#-G-G#-A-A#-B and back to C

```python
# Import necessary libraries
import librosa
import librosa.display as ld
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load sounds
cantina, rate = librosa.load('./slices/cantina.wav')
```

```python
cantina_chroma = librosa.feature.chroma_stft(cantina, sr = rate)
```

```python
fig, ax = plt.subplots(figsize=(25, 10), dpi=200)
a = ld.specshow(cantina_chroma, y_axis='chroma', x_axis='time', ax=ax, cmap='plasma')
fig.colorbar(a, ax=ax)
```

With this being a one second long clip, we do not get a huge amount of data to visualize. Even still, we can see that there is a large amount of stress on the D and A notes.

Let's try running this code on a longer wav file.

```python
star, rate = librosa.load('./slices/star.wav')
```

```python
star_chroma = librosa.feature.chroma_stft(star, sr = rate)
```

```python
fig, ax = plt.subplots(figsize=(30, 30), dpi=400)
a = ld.specshow(star_chroma, y_axis='chroma', x_axis='time', ax=ax, cmap='plasma')
fig.colorbar(a, ax=ax)
```

In this 60 second clip, we are able to get a strong understanding of which notes are frequented throughout the piece and can use that to infer which key we are in. Judging by the strong emphasis of F and B flat, we can guess with confidence that we are either in the key of F or B flat. It happens that we are in the key of B flat.
